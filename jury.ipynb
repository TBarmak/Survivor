# Jury Investigation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
overall = pd.read_excel('./Data/Leaderboard_ S1-S40.xlsx', sheet_name='PAE')
# Drop extra information below table
overall = overall.drop(overall.index[146:])
overall = overall.rename(columns={'single-': 'Season', 'contestant': 'Contestant'})

# Get rid of empty rows
overall = overall.dropna(subset=['Season', 'Contestant'])

# Reformat columns
overall['Season'] = overall['Season'].apply(lambda x: int(x[1:]))
overall['Finish'] = overall['Finish'].apply(lambda x: int(x))
overall['Time'] = overall['Time'].apply(lambda x: int(x))

# Add 'Sex' column from other sheet
full_seasons = pd.read_excel('./Data/Leaderboard_ S1-S40.xlsx')
full_seasons = full_seasons.rename(columns={'single-': 'Season', 'contestant': 'Contestant'})
full_seasons = full_seasons[['Season', 'Contestant', 'Sex']]
full_seasons['Season'] = full_seasons['Season'].apply(lambda x: int(x[1:]))
overall = pd.merge(full_seasons, overall, left_on=['Season', 'Contestant'], right_on=['Season', 'Contestant'])

# Sort by season and then by finish
overall.sort_values(['Season', 'Finish'], ascending=(True, True), inplace=True)
overall
```

```python
overall.columns
```

```python
# How many final 2s and Final 3s have there been
finals_by_season = overall.groupby('Season')['Contestant'].count()
print("Total number of final 2's vs final 3's:")
print(finals_by_season.value_counts())
overall['Final size'] = overall['Season'].apply(lambda x: finals_by_season[x])
```

```python
# Final 2 and final 3 over time
plt.scatter(finals_by_season.index, finals_by_season.values)
plt.show()
```

```python
from matplotlib.patches import Patch
# Percentage of final 2 vs final 3 over time
final_3_percent = [1 - overall.groupby('Season')['Contestant'].count().head(i).value_counts(normalize=True)[2] for i in finals_by_season.index]
plt.plot(final_3_percent)
plt.fill_between(np.arange(0, 40, 1), final_3_percent, color="red")
plt.fill_between(np.arange(0, 40, 1), final_3_percent, np.ones(40), color="blue")
plt.title("F2 vs F3 as Percent Frequency of Previous Seasons")
plt.xlabel("Season")
plt.ylabel("Percentage")
legend_elements = [Patch(facecolor='blue', edgecolor='b', label='F2'), Patch(facecolor='red', edgecolor='r', label='F3')]

plt.legend(handles=legend_elements, loc='upper left')
plt.show()
```

```python
# Most recent F2
print("Most recent F2 Season:", overall['Season'][overall['Final size'] == 2].tail(1).values[0])
```

```python
# Total people (not including repeats) who have sat at final tribal
len(overall.groupby('Contestant')['Finish'].mean())
```

```python
# Total number of jury votes cast all time
all_time_jury_votes = overall.groupby('Season')['TotJ'].mean().sum()
all_time_jury_votes
```

```python
# Total votes that went towards the winner
jury_votes_for_winner = overall['JVF'][overall['Finish'] == 1].sum()
print("Number of votes that went towards the winner:", jury_votes_for_winner)
print("Percentage:", "{:.2%}".format((jury_votes_for_winner / all_time_jury_votes)))
```

```python
# How many jury members are there per season
jurors_by_season = overall.groupby('Season')['TotJ'].mean()
print("Min number of jurors:", np.min(jurors_by_season))
print("Max number of jurors:", np.max(jurors_by_season))
print("Mean number of jurors:", np.mean(jurors_by_season))
print("Median number of jurors:", np.median(jurors_by_season))
```

```python
# Frequency of each configuration of final size and number of jury members
overall.groupby('Season')['Final size', 'TotJ'].mean().groupby(['Final size', 'TotJ']).size().unstack().fillna(0)
```

```python
# Scatter plot of jury members by season
plt.scatter(jurors_by_season.index, jurors_by_season.values)
```

```python
# Correlation between size of jury and how in agreement they are
winners = overall[:][overall["Finish"] == 1]
plt.scatter(winners['TotJ'], winners['JV%'])
plt.title("Fraction of Vote Earned by Winner vs People on Jury")
plt.xlabel("Number of Members on Jury")
plt.ylabel("Fraction of Vote Earned by Winner")
# No correlation
```

# Battle of Sexes Analysis

```python
# Sex of people sitting at final tribal
final_sex_dist = overall.groupby('Sex')['Contestant'].count()
print("Total males and females appearances at final tribal")
print(final_sex_dist)

# Removing duplicates
print("Removing duplicates...")
print("Total males and females who sat at final tribal")
overall.groupby('Contestant')['Sex'].min().value_counts()
```

```python
# Sex of winners by final size
winners.groupby(['Final size', 'Sex'])['Contestant'].count().unstack()
```

```python
from matplotlib.patches import Patch
# Percentage of male winners over time
male_winners_percent = [winners.groupby('Season')['Sex'].min().head(i).value_counts(normalize=True)[0] for i in range(1, 41)]
plt.plot(male_winners_percent)
plt.fill_between(np.arange(0, 40, 1), male_winners_percent, color="red")
plt.fill_between(np.arange(0, 40, 1), male_winners_percent, np.ones(40), color="blue")
plt.title("Male vs Female Winners as Percent Frequency of Previous Seasons")
plt.xlabel("Season")
plt.ylabel("Percentage")
legend_elements = [Patch(facecolor='blue', edgecolor='b', label='F'), Patch(facecolor='red', edgecolor='r', label='M')]

plt.legend(handles=legend_elements, loc='upper left')
plt.show()
```

```python
# Male/female combinations at final tribal and their frequencies
'''
-Most common tribal council has been 2 males and 1 female
-There have been 6 all female final tribal councils
-There have been 4 all male final tribal councils
'''
sex_counts_by_season = overall.groupby('Season')['Sex'].value_counts().unstack().fillna(0)
sex_counts_by_season.groupby(['M', 'F']).size()
```

```python
# Sex of winners where final tribal has male and female representation
winner_sexes = overall['Sex'][overall['Finish'] == 1]
winner_sexes.dropna(inplace=True)
winner_sexes.reset_index(drop=True, inplace=True)
sex_counts_by_season['Winner'] = list(winner_sexes)
sex_counts_by_season[:][(sex_counts_by_season['F'] > 0) & (sex_counts_by_season['M'] > 0)].groupby(['M', 'F', 'Winner']).size().unstack()
```
